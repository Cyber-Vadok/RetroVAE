```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
soil_img = 'biomless_tile/0.png'
void_img = 'biomless_tile/27.png'
```

```python
gray_soil = cv2.imread(soil_img)
gray_soil = cv2.cvtColor(gray_soil, cv2.COLOR_BGR2GRAY)

gray_void = cv2.imread(void_img)
gray_void = cv2.cvtColor(gray_void, cv2.COLOR_BGR2GRAY)
```

```python
# Plot the image using matplotlib
plt.imshow(gray_soil)
plt.axis('off')  # Hide the axis
plt.show()

# Plot the image using matplotlib
plt.imshow(gray_void)
plt.axis('off')  # Hide the axis
plt.show()
```
